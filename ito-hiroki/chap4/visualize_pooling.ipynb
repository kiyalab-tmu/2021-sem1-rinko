```python
from layer import MaxPooling, AvgPooling
import numpy as np
from PIL import Image
from IPython.core.display import display
```

```python
img = Image.open("goldfish.jpg")
```

```python
img
```

```python
img = np.array(img)
img.shape
```

```python
img_batch = np.expand_dims(img.transpose(2, 0, 1), 0)
img_batch.shape
```

## MaxPooling

```python
for size in [2, 5, 10, 20]:
    print(size)
    pooling = MaxPooling(size, size, stride=size, pad=0)
    pooled_img = Image.fromarray(pooling(img_batch).squeeze(axis=0).transpose(1, 2, 0).astype(np.uint8))
    display(pooled_img)
    print(pooled_img.size)
```

## AvgPooling

```python
for size in [2, 5, 10, 20]:
    print(size)
    pooling = AvgPooling(size, size, stride=size, pad=0)
    pooled_img = Image.fromarray(pooling(img_batch).squeeze(axis=0).transpose(1, 2, 0).astype(np.uint8))
    display(pooled_img)
    print(pooled_img.size)
```
