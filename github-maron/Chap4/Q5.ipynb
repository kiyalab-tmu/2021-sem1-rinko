```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import cv2

img = cv2.imread("kurumi.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

fig = plt.figure(figsize = (10, 10), constrained_layout = True)
ax1 = fig.add_subplot(1, 3, 1)
plt.title("original")
plt.imshow(img)

height, width, _ = img.shape

def pooling(x, image):
    ave_img = image.copy()
    max_img = image.copy()
    for i in range(x,height+x, x):
        for j in range(x,width+x, x):
            for k in range(3):
                fil = image[(i-x):i,(j-x):j,k]
                ave_img[(i-x):i,(j-x):j,k] = np.mean(fil)
                max_img[(i-x):i,(j-x):j,k] = np.max(fil)
    return ave_img, max_img



AVE, MAX = pooling(8, img)

ax2 = fig.add_subplot(1, 3, 2)
plt.imshow(AVE)
plt.title("average pooling")

ax1 = fig.add_subplot(1, 3, 3)
plt.imshow(MAX)
plt.title("max pooling")
```
