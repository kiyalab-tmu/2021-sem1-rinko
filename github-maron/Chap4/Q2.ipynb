```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import cv2

img = cv2.imread("ink.png")
RGB_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img = cv2.blur(img,(3,3))
width, height = img.shape
print(width, height)
plt.imshow(img, cmap = "gray")
```

```python
padding = np.zeros(((width+2, height+2)), dtype = int)

for i in range(width):
    for j in range(height):
        padding[i+1][j+1] = img[i][j]
```

```python
#1
padding1 = padding.copy()
ans = np.zeros(((width, height)), dtype = int)

H = [[1, 0, -1], [0, 0, 0],[-1, 0, 1]]

for i in range(width):
    for j in range(height):
        if np.sum((padding1[i:i+3,j:j+3]* H)) <= 0:
            ans[i][j] = 0
        else:
            ans[i][j] = 255
            
plt.imshow(ans, cmap = 'gray')
```

```python
#2
padding2 = padding.copy()
ans = np.zeros(((width, height)), dtype = int)

H = [[0, -1, 0], [-1, 4, -1],[0, -1, 0]]

for i in range(width):
    for j in range(height):
        if np.sum((padding2[i:i+3,j:j+3]* H)) <= 0:
            ans[i][j] = 0
        else:
            ans[i][j] = 255
            
plt.imshow(ans, cmap = 'gray')
```
